```python
# Mean estimate for spine density, based on the initial segment as the most reliable!


print('basal_mean_wg1')
print(np.mean(basal_first_wg1))
print('\n')

print('basal_mean_wg4')
print(np.mean(basal_first_wg4))
print('\n')

print('apical_mean_wg1')
print(np.mean(apical_first_wg1))
print('\n')

print('apical_mean_wg4')
print(np.mean(apical_first_wg4))
print('\n')

```

# plot all apicals

```python

fig, ax = plt.subplots()
group_1=[]
group_2=[]
group_1=apical_first_wg4
group_2=apical_first_wg1
plt.boxplot([group_2,group_1],labels=['WG1 (N='+str(len(group_2))+')','WG4 (N='+str(len(group_1))+')'])
#plt.violinplot([group_2,group_1])
d,p=stats.mannwhitneyu(group_1,group_2)
#d,p=stats.f_oneway(group_1,group_2)
print(p)
if p<0.05:
    plt.title('Apical dendrite initial segment spine density'+str(' (significant)') +str(' p= ') +str(np.round(p,8)))
else:
    plt.title('Apical dendrite initial segment spine density'+str(' (non-significant)') +str(' p= ') +str(np.round(p,8)))
plt.ylim([0, 1])
#plt.savefig('Apical_initial_segment.eps', format='eps', dpi=300)

fig, ax = plt.subplots()
group_1=[]
group_2=[]
group_1=apical_last_wg4
group_2=apical_last_wg1
plt.boxplot([group_2,group_1],labels=['WG1 (N='+str(len(group_2))+')','WG4 (N='+str(len(group_1))+')'])
#plt.violinplot([group_2,group_1])
d,p=stats.mannwhitneyu(group_1,group_2)
#d,p=stats.f_oneway(group_1,group_2)
print(p)
if p<0.05:
    plt.title('Apical dendrite last segment spine density'+str(' (significant)') +str(' p= ') +str(np.round(p,8)))
else:
    plt.title('Apical dendrite last segment spine density'+str(' (non-significant)') +str(' p= ') +str(np.round(p,8)))
plt.ylim([0, 1])
#plt.savefig('Apical_last_segment.eps', format='eps', dpi=300)

fig, ax = plt.subplots()
group_1=[]
group_2=[]
group_1=apical_total_wg4
group_2=apical_total_wg1
plt.boxplot([group_2,group_1],labels=['WG1 (N='+str(len(group_2))+')','WG4 (N='+str(len(group_1))+')'])
#plt.violinplot([group_2,group_1])
d,p=stats.mannwhitneyu(group_1,group_2)
#d,p=stats.f_oneway(group_1,group_2)
print(p)
if p<0.05:
    plt.title('Apical dendrite all spine density'+str(' (significant)') +str(' p= ') +str(np.round(p,8)))
else:
    plt.title('Apical dendrite all segment spine density'+str(' (non-significant)') +str(' p= ') +str(np.round(p,8)))
plt.ylim([0, 1])
#plt.savefig('Apical_all.eps', format='eps', dpi=300)
```
